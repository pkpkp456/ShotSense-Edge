```python
import os
import shutil
import pandas as pd

base_dir = '/content/drive/MyDrive/gunshot_fpga/data/raw'
gunshot_dir = os.path.join(base_dir, 'gunshot')
non_gunshot_dir = os.path.join(base_dir, 'non_gunshot')
os.makedirs(gunshot_dir, exist_ok=True)
os.makedirs(non_gunshot_dir, exist_ok=True)

print("Organizing datasets...")

# -------------------------------
# Helper function to safely move files
# -------------------------------
def safe_move(src_folder, dst_folder, prefix=""):
    for file in os.listdir(src_folder):
        src = os.path.join(src_folder, file)
        if os.path.isfile(src):
            dst_file = os.path.join(dst_folder, f"{prefix}_{file}" if prefix else file)
            shutil.move(src, dst_file)

# -------------------------------
# 1️⃣ UrbanSound8K
# -------------------------------
urbansound_audio_dir = os.path.join(gunshot_dir, 'UrbanSound8K/audio')
metadata_file = os.path.join(gunshot_dir, 'UrbanSound8K/metadata/UrbanSound8K.csv')

if os.path.exists(metadata_file):
    metadata = pd.read_csv(metadata_file)

    # Gunshot files
    gunshot_files = metadata[metadata['class'] == 'gun_shot']
    urbansound_gunshot_dir = os.path.join(gunshot_dir, 'urbansound8k')
    os.makedirs(urbansound_gunshot_dir, exist_ok=True)
    for _, row in gunshot_files.iterrows():
        fold = f"fold{row['fold']}"
        src = os.path.join(urbansound_audio_dir, fold, row['slice_file_name'])
        if os.path.exists(src):
            shutil.move(src, urbansound_gunshot_dir)

    # Non-gunshot files
    non_gunshot_files = metadata[metadata['class'] != 'gun_shot']
    urbansound_non_gunshot_dir = os.path.join(non_gunshot_dir, 'urbansound8k')
    os.makedirs(urbansound_non_gunshot_dir, exist_ok=True)
    for _, row in non_gunshot_files.iterrows():
        fold = f"fold{row['fold']}"
        src = os.path.join(urbansound_audio_dir, fold, row['slice_file_name'])
        if os.path.exists(src):
            safe_move([src], urbansound_non_gunshot_dir, prefix="urbansound8k")

    print("UrbanSound8K organized.")
else:
    print("UrbanSound8K metadata missing, skipping.")

# -------------------------------
# 2️⃣ Mendeley Gunshot
# -------------------------------
mendeley_dir = os.path.join(base_dir, 'x48cwz364j-3')
if os.path.exists(mendeley_dir):
    safe_move(mendeley_dir, gunshot_dir, prefix="mendeley")
    shutil.rmtree(mendeley_dir)
    print("Mendeley dataset organized safely.")

# -------------------------------
# 3️⃣ Zenodo Gunshot
# -------------------------------
zenodo_dir = os.path.join(base_dir, 'zenodo')
if os.path.exists(zenodo_dir):
    safe_move(zenodo_dir, gunshot_dir, prefix="zenodo")
    shutil.rmtree(zenodo_dir)
    print("Zenodo dataset organized safely.")

# -------------------------------
# 4️⃣ Edge-collected gunshot
# -------------------------------
edge_dir = os.path.join(gunshot_dir, 'edge_collected')
if os.path.exists(edge_dir):
    safe_move(edge_dir, gunshot_dir, prefix="edge")
    shutil.rmtree(edge_dir)
    print("Edge-collected gunshot dataset organized safely.")

# -------------------------------
# 5️⃣ ESC-50 → non-gunshot
# -------------------------------
esc_dir = os.path.join(non_gunshot_dir, 'ESC-50/audio')
if os.path.exists(esc_dir):
    safe_move(esc_dir, non_gunshot_dir, prefix="esc50")
    shutil.rmtree(os.path.join(non_gunshot_dir, 'ESC-50'))
    print("ESC-50 organized safely.")

# -------------------------------
# 6️⃣ MAD Dataset → non-gunshot
# -------------------------------
mad_dir = os.path.join(non_gunshot_dir, 'mad-dataset-military-audio-dataset')
if os.path.exists(mad_dir):
    safe_move(mad_dir, non_gunshot_dir, prefix="mad")
    shutil.rmtree(mad_dir)
    print("MAD dataset organized safely.")

print("✅ All datasets organized into gunshot and non_gunshot folders safely.")
```

```python
import os
import librosa
import numpy as np
import soundfile as sf
from tqdm import tqdm
import noisereduce as nr
from collections import defaultdict
import random

# === CONFIGURATION ===
RAW_DIR = "/content/drive/MyDrive/gunshot_fpga/data/raw"
PROCESSED_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed"

TARGET_SR = 32000
TARGET_DURATION = 2.0
TOP_DB = 20
USE_NOISE_REDUCTION = True

# Log-Mel parameters
N_MELS = 128
N_FFT = 1024
HOP_LENGTH = 320
FMIN = 100
FMAX = 8000
EPS = 1e-6

# Augmentation probability
AUG_PROB = 0.5

# === CREATE OUTPUT FOLDERS ===
os.makedirs(f"{PROCESSED_DIR}/gunshot", exist_ok=True)
os.makedirs(f"{PROCESSED_DIR}/non_gunshot", exist_ok=True)
os.makedirs(f"{PROCESSED_DIR}/logmel", exist_ok=True)

# === STATS TRACKERS ===
stats = defaultdict(lambda: {"count": 0, "durations": [], "errors": 0})
dataset_stats = defaultdict(lambda: defaultdict(int))

# === HELPER FUNCTIONS ===
def normalize_audio(y):
    return y / np.max(np.abs(y)) if np.max(np.abs(y)) != 0 else y

def trim_silence(y):
    yt, _ = librosa.effects.trim(y, top_db=TOP_DB)
    return yt

def reduce_noise(y, sr):
    try:
        return nr.reduce_noise(y=y, sr=sr, prop_decrease=0.75)
    except:
        return y

def pad_or_crop(y, sr, duration):
    target_len = int(duration * sr)
    if len(y) > target_len:
        y = y[:target_len]
    elif len(y) < target_len:
        y = np.pad(y, (0, target_len - len(y)))
    return y

def detect_dataset_name(file_path):
    parts = file_path.lower().split('/')
    if "urbansound8k" in parts:
        return "UrbanSound8K"
    elif "esc" in parts:
        return "ESC-50"
    elif "mad" in parts:
        return "MAD"
    elif "mivia" in parts:
        return "MIVIA"
    elif "edge" in parts:
        return "Edge-Collected"
    elif "gunshot" in parts and "dataset" in parts:
        return "GunshotAudioDataset"
    elif "x48cwz364j" in parts:
        return "Mendeley"
    else:
        return "Unknown"

def extract_log_mel(y, sr):
    mel = librosa.feature.melspectrogram(
        y=y,
        sr=sr,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        fmin=FMIN,
        fmax=FMAX
    )
    log_mel = np.log(mel + EPS)
    log_mel = (log_mel - np.mean(log_mel)) / (np.std(log_mel) + EPS)
    return log_mel

def augment_audio(y, sr):
    # Random pitch shift ±2 semitones
    if random.random() < AUG_PROB:
        n_steps = random.uniform(-2, 2)
        y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=n_steps)

    # Random time stretch ±10%
    if random.random() < AUG_PROB:
        rate = random.uniform(0.9, 1.1)
        y = librosa.effects.time_stretch(y=y, rate=rate)
        y = pad_or_crop(y, sr, TARGET_DURATION)

    # Random time shift ±50 ms
    if random.random() < AUG_PROB:
        shift = int(random.uniform(-0.05, 0.05) * sr)
        y = np.roll(y, shift)

    return y

def preprocess_audio(file_path, out_dir_wav, out_dir_logmel, label):
    try:
        y, sr = librosa.load(file_path, sr=TARGET_SR, mono=True)
        y = trim_silence(y)
        y = normalize_audio(y)
        if USE_NOISE_REDUCTION:
            y = reduce_noise(y, sr)
        y = pad_or_crop(y, sr, TARGET_DURATION)
        y = augment_audio(y, sr)

        # Save WAV (overwrite if exists)
        base_name = os.path.splitext(os.path.basename(file_path))[0]
        dataset_name = detect_dataset_name(file_path)
        save_name_wav = f"{dataset_name}_{label}_{base_name}.wav"
        out_path_wav = os.path.join(out_dir_wav, save_name_wav)
        sf.write(out_path_wav, y, sr)

        # Extract and save log-Mel
        log_mel = extract_log_mel(y, sr)
        save_name_logmel = save_name_wav.replace(".wav", ".npy")
        out_path_logmel = os.path.join(out_dir_logmel, save_name_logmel)
        np.save(out_path_logmel, log_mel)

        # Update stats
        stats[label]["count"] += 1
        stats[label]["durations"].append(len(y) / sr)
        dataset_stats[label][dataset_name] += 1

    except Exception as e:
        stats[label]["errors"] += 1
        print(f"❌ Error processing {file_path}: {e}")

# === RUN PREPROCESSING ===
for label in ["gunshot", "non_gunshot"]:
    input_dir = os.path.join(RAW_DIR, label)
    output_dir_wav = os.path.join(PROCESSED_DIR, label)
    output_dir_logmel = os.path.join(PROCESSED_DIR, "logmel", label)
    os.makedirs(output_dir_logmel, exist_ok=True)

    all_files = []
    for root, _, files in os.walk(input_dir):
        for f in files:
            if f.lower().endswith(".wav"):
                all_files.append(os.path.join(root, f))

    print(f"\n🔄 Processing {label.upper()} ({len(all_files)} files found)...")

    for file_path in tqdm(all_files):
        preprocess_audio(file_path, output_dir_wav, output_dir_logmel, label)

# === SUMMARY REPORT ===
print("\n📊 === DETAILED PREPROCESSING SUMMARY ===")
for label, info in stats.items():
    print(f"\n📁 CATEGORY: {label.upper()}")
    print(f"✅ Processed Files: {info['count']}")
    print(f"⚠️ Errors Skipped: {info['errors']}")
    avg_duration = np.mean(info['durations']) if info['durations'] else 0
    print(f"⏱️ Avg. Duration: {avg_duration:.2f} sec")
    print("📦 Per-Dataset Breakdown:")
    for dataset, count in dataset_stats[label].items():
        print(f"   - {dataset}: {count} files")
    example_files = os.listdir(os.path.join(PROCESSED_DIR, "logmel", label))[:3]
    print(f"📄 Example Log-Mel Outputs: {example_files}")

print("\n✅ All audio processed, augmented, and log-Mel spectrograms saved successfully!")
```

```python
import os
import librosa
import numpy as np
from scipy.signal import find_peaks
from tqdm import tqdm
import shutil

# Folders
RAW_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/gunshot"
FILTERED_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/gunshot_filtered"
os.makedirs(FILTERED_DIR, exist_ok=True)

# Parameters
POWER_THRESHOLD = 0.002
PEAK_FACTOR = 1.2
MIN_PEAK_COUNT = 1

# Helper functions
def compute_rms(y):
    return np.mean(y ** 2)

def count_rms_peaks(y, frame_length=512, hop_length=256):
    rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]
    threshold = np.median(rms) * PEAK_FACTOR
    peaks, _ = find_peaks(rms, height=threshold)
    return len(peaks)

# Filter all gunshot files
files = [f for f in os.listdir(RAW_DIR) if f.endswith(".wav")]
kept_files = 0

for f in tqdm(files):
    filepath = os.path.join(RAW_DIR, f)
    y, sr = librosa.load(filepath, sr=16000, mono=True)

    power = compute_rms(y)
    peak_count = count_rms_peaks(y)

    if power >= POWER_THRESHOLD and peak_count >= MIN_PEAK_COUNT:
        out_path = os.path.join(FILTERED_DIR, f)
        # Overwrite if file already exists
        shutil.copyfile(filepath, out_path)
        kept_files += 1

print(f"\n✅ Light RMS filtering with overwrite completed. Kept {kept_files}/{len(files)} files.")
```

```python
import os
import shutil
from collections import defaultdict
from tqdm import tqdm
import math

# Paths
NON_GUNSHOT_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/non_gunshot"
BALANCED_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/non_gunshot_balanced"
os.makedirs(BALANCED_DIR, exist_ok=True)

TARGET_COUNT = 5773  # Match gunshot count

# --- Step 1: Organize files by category ---
category_files = defaultdict(list)

for f in os.listdir(NON_GUNSHOT_DIR):
    if f.endswith(".wav"):
        # Example: filenames like "dog_bark_001.wav" -> category = "dog_bark"
        category = "_".join(f.split("_")[:-1])
        category_files[category].append(f)

# --- Step 2: Compute proportional selection per category ---
num_categories = len(category_files)
files_per_category = math.ceil(TARGET_COUNT / num_categories)

selected_files = []

for category, files in category_files.items():
    # Shuffle files within category
    files_sorted = sorted(files)  # optional: sort for reproducibility
    # Select min(files_per_category, total available)
    selected = files_sorted[:min(files_per_category, len(files))]
    selected_files.extend(selected)

# If we got more than TARGET_COUNT, truncate
selected_files = selected_files[:TARGET_COUNT]

# --- Step 3: Copy selected files ---
for f in tqdm(selected_files):
    src = os.path.join(NON_GUNSHOT_DIR, f)
    dst = os.path.join(BALANCED_DIR, f)
    shutil.copyfile(src, dst)

print(f"\n✅ Category-based balancing completed. Selected {len(selected_files)} files.")
```

```python
import os
import librosa
import numpy as np
from tqdm import tqdm
import soundfile as sf

# =========================
# PARAMETERS
# =========================
SR = 32000            # Sample rate
WINDOW_DURATION = 2.0 # seconds
STRIDE_DURATION = 1.0 # seconds (overlap)
EPS = 1e-6

# =========================
# PATHS
# =========================
FILTERED_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/gunshot_filtered"
BALANCED_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/non_gunshot_balanced"

SLIDING_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/sliding_windows"
os.makedirs(SLIDING_DIR, exist_ok=True)

DATASETS = {
    "gunshot": FILTERED_DIR,
    "non_gunshot": BALANCED_DIR
}

# =========================
# FUNCTIONS
# =========================
def sliding_window_audio(y, sr=SR, window_duration=WINDOW_DURATION, stride_duration=STRIDE_DURATION):
    """Slice audio into overlapping windows"""
    window_len = int(window_duration * sr)
    stride_len = int(stride_duration * sr)
    segments = []
    for start in range(0, len(y) - window_len + 1, stride_len):
        segments.append(y[start:start+window_len])
    return segments

def process_and_save_windows(input_dir, output_dir, sr=SR):
    os.makedirs(output_dir, exist_ok=True)
    files = librosa.util.find_files(input_dir, ext=['wav'])
    count = 0
    for f in tqdm(files):
        y, _ = librosa.load(f, sr=sr, mono=True)
        segments = sliding_window_audio(y, sr)
        for i, seg in enumerate(segments):
            base = os.path.splitext(os.path.basename(f))[0]
            out_path = os.path.join(output_dir, f"{base}_win{i}.wav")
            sf.write(out_path, seg, sr)
            count += 1
    print(f"✅ Processed {count} sliding-window segments from {input_dir}")

# =========================
# RUN
# =========================
for label, folder in DATASETS.items():
    print(f"\n🔹 Processing {label.upper()} dataset into sliding windows...")
    output_subdir = os.path.join(SLIDING_DIR, label)
    process_and_save_windows(folder, output_subdir)
```

```python
import os
import librosa
import numpy as np
from tqdm import tqdm

# =========================
# PARAMETERS (PANNs CNN14 compatible)
# =========================
SR = 32000                # Sampling rate
N_FFT = 1024              # FFT window size (~32 ms)
HOP_LENGTH = 320          # 10 ms hop
N_MELS = 64               # Number of Mel bands
FMIN, FMAX = 50, 14000    # Frequency range of interest
EPS = 1e-6                # Numerical stability

# =========================
# PATHS
# =========================
SLIDING_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/sliding_windows"
FEATURE_DIR = "/content/drive/MyDrive/gunshot_fpga/data/features_logmel_sliding"
os.makedirs(FEATURE_DIR, exist_ok=True)

DATASETS = {
    "gunshot": os.path.join(SLIDING_DIR, "gunshot"),
    "non_gunshot": os.path.join(SLIDING_DIR, "non_gunshot")
}

# =========================
# FUNCTIONS
# =========================
def extract_logmel(y, sr=SR):
    """Compute log-Mel spectrogram and normalize per frequency"""
    mel = librosa.feature.melspectrogram(
        y=y,
        sr=sr,
        n_fft=N_FFT,
        hop_length=HOP_LENGTH,
        n_mels=N_MELS,
        fmin=FMIN,
        fmax=FMAX,
        window='hann'
    )
    logmel = np.log(mel + EPS)
    # Per-frequency normalization
    mean = np.mean(logmel, axis=1, keepdims=True)
    std = np.std(logmel, axis=1, keepdims=True)
    logmel_norm = (logmel - mean) / (std + EPS)
    return logmel_norm

# =========================
# MAIN EXTRACTION LOOP
# =========================
for label, folder in DATASETS.items():
    print(f"\n🎧 Extracting log-Mel spectrograms for {label.upper()} dataset...")

    output_subdir = os.path.join(FEATURE_DIR, label)
    os.makedirs(output_subdir, exist_ok=True)

    files = librosa.util.find_files(folder, ext=['wav'])
    print(f"Found {len(files)} sliding-window files.")

    for f in tqdm(files):
        try:
            y, _ = librosa.load(f, sr=SR, mono=True)
            logmel = extract_logmel(y)

            base = os.path.splitext(os.path.basename(f))[0]
            np.save(os.path.join(output_subdir, base + ".npy"), logmel)
        except Exception as e:
            print(f"❌ Error with {f}: {e}")

print("\n✅ Log-Mel spectrogram extraction completed successfully for all sliding windows.")
```

```python
import torch
import torch.nn as nn

# BiLSTM + Attention Head
class BiLSTM_Attention(nn.Module):
    def __init__(self, emb_dim=2048, hidden_dim=128, num_classes=2):
        super().__init__()
        self.bilstm = nn.LSTM(
            input_size=emb_dim,
            hidden_size=hidden_dim,
            num_layers=1,
            batch_first=True,
            bidirectional=True
        )
        self.attention = nn.Sequential(
            nn.Linear(hidden_dim*2, 128),
            nn.Tanh(),
            nn.Linear(128, 1),
            nn.Softmax(dim=1)
        )
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim*2, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        lstm_out, _ = self.bilstm(x)
        attn_weights = self.attention(lstm_out)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

# Full pipeline: CNN14 backbone + BiLSTM + Attention
class GunshotModel(nn.Module):
    def __init__(self, cnn14, lstm_hidden=128, num_classes=2):
        super().__init__()
        self.cnn14 = cnn14
        self.head = BiLSTM_Attention(emb_dim=2048, hidden_dim=lstm_hidden, num_classes=num_classes)

    def forward(self, x):
        if x.ndim == 5:
            x = x.squeeze(2)  # remove extra dimension
        emb = self.cnn14(x)['embedding']  # (B, 2048)
        emb_seq = emb.unsqueeze(1)        # (B, 1, 2048) -> LSTM expects 3D
        return self.head(emb_seq)
```

```python
import torch
from torch.utils.data import Dataset
import numpy as np

class GunshotDataset(Dataset):
    def __init__(self, samples, min_frames=256):
        """
        samples: list of tuples (file_path, label)
        min_frames: minimum number of time frames for CNN14
        """
        self.samples = samples
        self.min_frames = min_frames

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        x = np.load(path)  # shape (n_mels, frames) e.g., (64, 201)
        x = torch.tensor(x, dtype=torch.float32)
        x = x.transpose(0, 1).unsqueeze(0)  # (1, frames, n_mels)

        # =========================
        # Pad time dimension if too short
        # =========================
        T = x.shape[1]
        if T < self.min_frames:
            pad_len = self.min_frames - T
            # pad last dimension of time (dim=1)
            x = torch.nn.functional.pad(x, (0, 0, 0, pad_len))

        return x, torch.tensor(label, dtype=torch.long)
```

```python
import os
import torch
import librosa
import numpy as np
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Directories
RAW_AUDIO_DIR = "/content/drive/MyDrive/gunshot_fpga/data/processed/sliding_windows"
EMB_DIR = "/content/drive/MyDrive/gunshot_fpga/data/embeddings_audio"
os.makedirs(EMB_DIR, exist_ok=True)

# CNN14 is already loaded as cnn14
cnn14.eval()
for param in cnn14.parameters():
    param.requires_grad = False

# Parameters
SR = 32000  # CNN14 sample rate

for label in ['gunshot', 'non_gunshot']:
    folder = os.path.join(RAW_AUDIO_DIR, label)
    output_folder = os.path.join(EMB_DIR, label)
    os.makedirs(output_folder, exist_ok=True)
    files = [f for f in os.listdir(folder) if f.endswith('.wav')]

    for f in tqdm(files, desc=f"Processing {label}"):
        # Load audio
        y, _ = librosa.load(os.path.join(folder, f), sr=SR, mono=True)
        y_tensor = torch.tensor(y, dtype=torch.float32).unsqueeze(0).to(device)  # (1, samples)

        # Forward pass to get embedding
        with torch.no_grad():
            emb = cnn14(y_tensor)['embedding']  # (1, 2048)

        emb = emb.squeeze(0).cpu().numpy()
        np.save(os.path.join(output_folder, f.replace('.wav', '.npy')), emb)
```

```python
import os
import torch
import numpy as np
import librosa
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, Dataset
import torch.nn as nn
import torch.optim as optim
```

```python
from panns_inference.models import Cnn14

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cnn14 = Cnn14(
    sample_rate=32000,
    window_size=1024,
    hop_size=320,
    mel_bins=64,
    fmin=50,
    fmax=14000,
    classes_num=527
).to(device)

cnn14.eval()
for p in cnn14.parameters():
    p.requires_grad = False

state_dict = torch.load("/content/drive/MyDrive/gunshot_fpga/data/models/Cnn14_mAP=0.431.pth", map_location=device, weights_only=False)
cnn14.load_state_dict(state_dict['model'])
print("✅ CNN14 loaded successfully.")
```

```python
EMB_DIR = "/content/drive/MyDrive/gunshot_fpga/data/embeddings_audio"

samples = []
for label in ['gunshot', 'non_gunshot']:
    folder = os.path.join(EMB_DIR, label)
    files = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith('.npy')]
    for f in files:
        samples.append((f, 1 if label=='gunshot' else 0))

train_files, test_files = train_test_split(samples, test_size=0.15, random_state=42, stratify=[s[1] for s in samples])
train_files, val_files  = train_test_split(train_files, test_size=0.1765, random_state=42, stratify=[s[1] for s in train_files])

class GunshotAudioEmbeddingDataset(Dataset):
    def __init__(self, samples):
        self.samples = samples
    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        path, label = self.samples[idx]
        x = np.load(path)
        x = torch.tensor(x, dtype=torch.float32).unsqueeze(0)  # (1, 2048)
        return x, torch.tensor(label, dtype=torch.long)

batch_size = 32
train_dl = DataLoader(GunshotAudioEmbeddingDataset(train_files), batch_size=batch_size, shuffle=True, num_workers=2)
val_dl   = DataLoader(GunshotAudioEmbeddingDataset(val_files), batch_size=batch_size, shuffle=False, num_workers=2)
test_dl  = DataLoader(GunshotAudioEmbeddingDataset(test_files), batch_size=batch_size, shuffle=False, num_workers=2)

print(f"✅ DataLoaders ready: {len(train_dl.dataset)} train, {len(val_dl.dataset)} val, {len(test_dl.dataset)} test samples")
```

```python
import torch.nn as nn

class BiLSTM_Attention(nn.Module):
    def __init__(self, emb_dim=2048, hidden_dim=128, num_classes=2):
        super().__init__()
        self.bilstm = nn.LSTM(input_size=emb_dim, hidden_size=hidden_dim, batch_first=True, bidirectional=True)
        self.attention = nn.Sequential(
            nn.Linear(hidden_dim*2, 128),
            nn.Tanh(),
            nn.Linear(128, 1),
            nn.Softmax(dim=1)
        )
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim*2, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        lstm_out, _ = self.bilstm(x)
        attn_weights = self.attention(lstm_out)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

model = BiLSTM_Attention(emb_dim=2048, hidden_dim=128, num_classes=2).to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
num_epochs = 50
checkpoint_dir = "/content/drive/MyDrive/gunshot_fpga/models_checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)
best_val_acc = 0.0

for epoch in range(1, num_epochs+1):
    model.train()
    running_loss, correct, total = 0.0, 0, 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(x)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * x.size(0)
        preds = torch.argmax(out, dim=1)
        correct += (preds == y).sum().item()
        total += y.size(0)

    train_loss = running_loss / total
    train_acc = correct / total * 100

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for x_val, y_val in val_dl:
            x_val, y_val = x_val.to(device), y_val.to(device)
            out_val = model(x_val)
            loss_val = criterion(out_val, y_val)
            val_loss += loss_val.item() * x_val.size(0)
            preds_val = torch.argmax(out_val, dim=1)
            val_correct += (preds_val == y_val).sum().item()
            val_total += y_val.size(0)

    val_loss /= val_total
    val_acc = val_correct / val_total * 100

    print(f"Epoch {epoch}/{num_epochs} | "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}% | "
          f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), os.path.join(checkpoint_dir, f"best_model_epoch{epoch}.pth"))
        print(f"✅ Saved new best model checkpoint")
```

```python
import torch
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Put model in evaluation mode
model.eval()

def evaluate_model(model, dataloader, device, split_name="Validation"):
    all_preds = []
    all_labels = []
    total_loss = 0.0
    criterion = torch.nn.CrossEntropyLoss()

    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            outputs = model(x)
            loss = criterion(outputs, y)
            total_loss += loss.item() * x.size(0)

            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(y.cpu().numpy())

    avg_loss = total_loss / len(dataloader.dataset)
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    accuracy = (all_preds == all_labels).mean() * 100

    print(f"\n📊 {split_name} Results:")
    print(f"  ➤ Loss: {avg_loss:.4f}")
    print(f"  ➤ Accuracy: {accuracy:.2f}%")
    print("\nDetailed classification report:")
    print(classification_report(all_labels, all_preds, target_names=["Non-Gunshot", "Gunshot"]))

    print("Confusion Matrix:")
    print(confusion_matrix(all_labels, all_preds))

    return avg_loss, accuracy, all_labels, all_preds


# Run evaluation
val_loss, val_acc, val_labels, val_preds = evaluate_model(model, val_dl, device, split_name="Validation")
test_loss, test_acc, test_labels, test_preds = evaluate_model(model, test_dl, device, split_name="Test")
```
